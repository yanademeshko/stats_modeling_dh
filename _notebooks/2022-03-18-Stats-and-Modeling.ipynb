# Stats and Modeling

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
with open('Macbeth Shakespeare.txt') as f:
    x = f.read()
```

### Tokenizing words

```python
sent = sent_tokenize(x)
print(sent)
```

```python
print(word_tokenize(sent[1]))
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

### Removing stopwords

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
extrastop = ['’', '“','”']
```

```python
print(punctuation)
```

```python
print(stopwords.words('english'))
```

```python
print(list(punctuation))
```

```python
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
```

```python
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
print(wordsNoStop)
```

```python
wordsNoStopComp = [w for w in words if w not in myStopWords]
print(wordsNoStopComp)
```

### Stemming/Lemmatizing words

```python
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.porter import PorterStemmer
```

```python
wordLancasterStems = [LancasterStemmer().stem(w) for w in wordsNoStopComp]
```

```python
print(wordLancasterStems)
```

### Calculating word frequences

```python
from nltk.corpus import wordnet
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordLancasterStems)
```

```python
freq
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(freq.items()), columns = ["Word", "Frequency"])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = features_df[['bmi']]
y = features_df['target_column']
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,y,'ro',X,ytrain,'b-');
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```
